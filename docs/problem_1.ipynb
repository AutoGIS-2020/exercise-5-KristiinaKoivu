```python
# Import modules
import rasterio
from rasterio.plot import show
import numpy as np
from matplotlib import pyplot

# Open the raster file in read mode 
fp_11 = "tilavuus_11.tif"
fp_17 = "tilavuus_17.tif"
raster11 = rasterio.open(fp_11)
raster17 = rasterio.open(fp_17)

# Read in the data from raster files in band 1 
vol11 = raster11.read()
vol17 = raster17.read()

# Convert to floats
#vol11 = vol11.astype('f4')
#vol17 = vol17.astype('f4')

#np.seterr(divide='ignore', invalid='ignore')

vol11 = np.where(vol11 == np.nan, 0, vol11)
vol17 = np.where(vol17 == np.nan, 0, vol17)

# Check the data
print(vol11)
print(vol17)
print(type(vol11))
print(type(vol17))


show(vol11, cmap="terrain")
show(vol17, cmap="terrain")
```

```python
print(vol11.ndim)
print(vol17.ndim)

print(vol11.shape)
print(vol17.shape)

print(vol11.size)
print(vol17.size)
```
